```python
import pandas as pd
def load_data(xtest, ytest):
    test_data = pd.read_csv(xtest, header=None).values.flatten()
    true_labels = pd.read_csv(ytest, header=None).values.flatten()
    return test_data, true_labels

curr_dir = sys.path[0]
print (curr_dir)

# Calculate accuracy
def calculate_accuracy(predictions, true_labels):
    correct_predictions = np.sum(predictions == true_labels)
    accuracy = correct_predictions / len(true_labels)
    return accuracy

# Path to your CSV files
input_csv = os.path.join(curr_dir, 'xtest.csv')
labels_csv = os.path.join(curr_dir, 'ytest.csv')

# Load the test data and labels
test_data, true_labels = load_data(input_csv, labels_csv)

print(test_data)
```

```python

# Convert test data to integers
int_data = np.array([float_to_int(f) for f in test_data], dtype=np.uint32)

chunk_size = 144
num_chunks = len(test_data) // chunk_size
assert len(test_data) % chunk_size == 0, "The length of test data should be divisible by chunk size."

# Allocate DMA buffers
input_buffer = allocate(shape=(144,), dtype=np.uint32)
output_buffer = allocate(shape=(1,), dtype=np.uint32)

all_predictions = []

# Process data in chunks
start_time = time.time()
for i in range(num_chunks):
    # Extract chunk
    chunk_start = i * chunk_size
    chunk_end = chunk_start + chunk_size
    chunk_data = test_data[chunk_start:chunk_end]

    # Convert chunk data to integers
    int_data = np.array([float_to_int(f) for f in chunk_data], dtype=np.uint32)

    # Copy data to input buffer
    np.copyto(input_buffer, int_data)

    # Ensure output buffer is initialized to zero
    output_buffer[:] = 0

    # Start DMA transfer
    dma_send.transfer(input_buffer)
    dma_recv.transfer(output_buffer)

    # Wait for DMA operations to complete
    dma_send.wait()
    dma_recv.wait()

    # Convert received output from integers to floats
    predictions_chunk = np.array(output_buffer)

    # Round predictions to nearest integer if necessary
    predictions_chunk = np.round(predictions_chunk)

    # Store predictions
    all_predictions.extend(predictions_chunk)

    
end_time = time.time()
# Calculate accuracy
all_predictions = np.array(all_predictions)
accuracy = calculate_accuracy(all_predictions, true_labels)



# Print results
print("Predictions:", all_predictions)
print("True Labels:", true_labels)
print("Accuracy:", accuracy)
print('Time taken:', end_time - start_time)
```
